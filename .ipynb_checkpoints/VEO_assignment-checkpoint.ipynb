## Player with most rise and drop in potential

```python
# Query with join between both tables and limit to the top 3 results in order to see if a tie is present

player_potential_rise = pd.read_sql('''
    SELECT d19.ID,d19.Name,d21.team, d19.Potential AS potential19,
    d21.potential AS potential21, d21.potential - d19.Potential AS potential_dif 
    FROM data2019 AS d19 
    INNER JOIN data2021 AS d21 
    ON d19.ID = d21.player_id
    ORDER BY potential_dif DESC
    LIMIT 3
''', con)
```

```python
player_potential_rise
```

```python
# Query with join between both tables and limit to the top 3 results in order to see if a tie is present as in this case

player_potential_drop = pd.read_sql('''
    SELECT d19.ID,d19.Name,d21.team, d19.Potential AS potential19,
    d21.potential AS potential21, d21.potential - d19.Potential AS potential_dif 
    FROM data2019 AS d19 
    INNER JOIN data2021 AS d21 
    ON d19.ID = d21.player_id
    ORDER BY potential_dif 
    LIMIT 3
''', con)
```

```python
player_potential_drop
```

```python
# Query with CTE and returning min and max value

prd = pd.read_sql('''
    WITH rise_drop (ID,Name,Team,potential19,potential21,potential_dif) AS(
    SELECT d19.ID,d19.Name,d21.team, d19.Potential AS potential19,
    d21.potential AS potential21, d21.potential - d19.Potential AS potential_dif
    FROM data2019 AS d19 
    INNER JOIN data2021 AS d21 
    ON d19.ID = d21.player_id
    )
    SELECT *
    FROM rise_drop
    WHERE potential_dif = (SELECT MIN(potential_dif) FROM rise_drop)
    UNION ALL
    SELECT *
    FROM rise_drop
    WHERE potential_dif = (SELECT MAX(potential_dif) FROM rise_drop)
''', con)
```

```python
prd
```

## New players in 21 and players present in 19, not present in 21

```python
# Query with filter for players present only in 2021

new_players = pd.read_sql('''
    SELECT count(player_id) AS new_players
    FROM data2021
    WHERE player_id NOT IN (SELECT ID FROM data2019)
''', con)
```

```python
new_players
```

```python
# Query with filter for players present in 19, not present in 21

old_players = pd.read_sql('''
    SELECT count(ID) AS old_players
    FROM data2019
    WHERE ID NOT IN (SELECT player_id FROM data2021)
''', con)
```

```python
old_players
```

```python
# Query - Countries with most old players

countries_with_most_old_players = pd.read_sql('''
    SELECT nationality,count(ID) AS old_players_count
    FROM data2019
    WHERE ID NOT IN (SELECT player_id FROM data2021)
    GROUP BY nationality
    ORDER BY old_players_count DESC
    LIMIT 10
''', con)

fig1 = px.bar(countries_with_most_old_players,x='Nationality', y='old_players_count')
fig1.update_layout(legend=dict(x=0.75, y=1),
                  title_text="Top 10 countries with most old players",
                  title_x=0.5, xaxis_title="Nationality",
                  yaxis_title="Count",paper_bgcolor='lightgoldenrodyellow',
                 width=850, height=450, plot_bgcolor='lightgoldenrodyellow')
```

```python
# Query - Countries with most old players

countries_with_most_new_players = pd.read_sql('''
    SELECT nationality,count(player_id) AS new_players
    FROM data2021
    WHERE player_id NOT IN (SELECT ID FROM data2019)
    GROUP BY nationality
    ORDER BY new_players DESC
    LIMIT 10
''', con)

fig1 = px.bar(countries_with_most_new_players,x='nationality', y='new_players')
fig1.update_layout(legend=dict(x=0.75, y=1),
                  title_text="Top 10 countries with most new players",
                  title_x=0.5, xaxis_title="Nationality",
                  yaxis_title="Count",paper_bgcolor='lightgoldenrodyellow',
                 width=850, height=450, plot_bgcolor='lightgoldenrodyellow')
```

```python
# Query for nationality distribution of the 2021 players, per country

players_per_country = pd.read_sql('''
    SELECT nationality,count(player_id) AS n_players
    FROM data2021
    GROUP BY nationality
    ORDER BY n_players DESC
''', con)
```

```python
players_per_country
```

```python
# Query for nationality distribution of the 2021 players, per country top 10

players_per_country_top10 = pd.read_sql('''
    SELECT nationality,count(player_id) AS n_players
    FROM data2021
    GROUP BY nationality
    ORDER BY n_players DESC
    LIMIT 10
''', con)

fig2 = px.bar(players_per_country_top10, x='nationality', y='n_players')
fig2.update_layout(legend=dict(x=0.75, y=1),
                  title_text="Top 10 countries with most players",
                  title_x=0.5, xaxis_title="Nationality",
                  yaxis_title="Count",paper_bgcolor='lightgoldenrodyellow',
                 width=850, height=450, plot_bgcolor='lightgoldenrodyellow')
```

```python
# Query with CTE to find the distribution of overall performance for players from the top 10 countries with most players

top_10_overall = pd.read_sql('''
    WITH top_10(nationality,id) AS
    (SELECT nationality,count(player_id) AS n_players
    FROM data2021
    GROUP BY nationality
    ORDER BY n_players DESC
    LIMIT 10)
    SELECT nationality,overall
    FROM data2021
    WHERE nationality IN (SELECT nationality FROM top_10)
''', con)
```

```python
top_10_overall
```

```python
fig3 = px.box(top_10_overall, x='nationality',y='overall')
fig3.update_layout(legend=dict(x=0.75, y=1),
                  title_text="England is with most avarage players in FIFA2021",
                  title_x=0.5, xaxis_title="Nationality",
                  yaxis_title="Overall",paper_bgcolor='lightgoldenrodyellow',
                 width=850, height=450, plot_bgcolor='lightgoldenrodyellow')
```
